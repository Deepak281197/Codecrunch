```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold,LeaveOneOut,cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
my_data= load_iris()
```

```python
my_data
```

```python
features= pd.DataFrame(my_data.data,columns=my_data.feature_names)
target= pd.DataFrame(my_data.target, columns=['Target'])
target
```

```python
leave= LeaveOneOut()
```

```python
for train_split, test_split in leave.split(features):
    print(train_split,test_split)
```

```python
kfold= KFold(n_splits=5,shuffle=True)
```

```python
for train_split, test_split in kfold.split(features):
    print(train_split,test_split)
```

```python
rand_forest= RandomForestClassifier(n_estimators=10,bootstrap=True)
```

```python
cross_val_score(rand_forest,features,target,scoring='accuracy',cv=kfold,n_jobs=-1).mean()
```

## Survival Analysis

```python
from lifelines import KaplanMeierFitter
```

```python
df= pd.read_csv('survival_unemployment.csv')
```

```python
df.head()
```

```python
time_period= df['spell']
```

```python
time_period
```

```python
kaplan= KaplanMeierFitter()
```

```python
kaplan.fit(time_period,df.event)
kaplan.plot()
```

```python
kaplan.fit(time_period[df.ui==0],df.event[df.ui==0],label='0')
kaplan.plot()
kaplan.fit(time_period[df.ui==1],df.event[df.ui==1],label='1')
kaplan.plot()
```
