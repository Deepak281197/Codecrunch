# CrossTab

```python
a
```

```python
a.columns
```

```python
pd.crosstab(index=a['name_of_car'],columns=a['Model'])
```

```python
pd.crosstab(index=[a['name_of_car'],a['Type']],columns=[a['EngineSize'],a['Weight']])
```

```python
pd.crosstab(index=[a['name_of_car'],a['Type']],columns=[a['EngineSize'],a['Weight']], margins= True)
```

# Groupby

```python
a
```

```python
a['Origin'].value_counts()
```

```python
a.groupby(a['name_of_car']).get_group('Toyota')
```

```python
a.groupby([a['name_of_car'],a['Origin'],a['Type']])['Weight','EngineSize'].min()
```

```python
import numpy as np
a.groupby([a['name_of_car'],a['Type'],a['Origin']])['Weight','EngineSize'].agg([max,min,np.mean])
```

```python
a[['Weight','EngineSize']].groupby(a['name_of_car']).agg([min,max])
```
