```python
import spacy
```

```python
nlp= spacy.load('en_core_web_sm')
```

```python
doc2=nlp(doc1)
```

```python
for token in doc2:
  print(token,token.pos_,token.lemma_)
```

```python
noun_words= [token.text for token in doc2 if token.pos_ in ['NOUN','VERB','ADJ']]
print(noun_words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
ct= CountVectorizer()
```

```python
x= ct.fit_transform(noun_words)
```

```python
print(ct.vocabulary_.items())
```

```python
x.shape
```

```python
x.toarray()
```

```python
sum_x= x.sum(axis=0)
sum_x
```

```python
sum_x.shape
```

```python
for i,j in ct.vocabulary_.items():
  print(i,sum_x[0,j])
```

```python
word_freq= [(words,sum_x[0,idf]) for words, idf in ct.vocabulary_.items()]
word_freq= sorted(word_freq, reverse=True, key= lambda x:x[1])
word_df= pd.DataFrame(word_freq, columns=['word','count'])
word_df[0:10]
```

```python
word_df[0:10].plot(kind='bar',x='word',title='Top Repeated words',figsize=(10,6))
```

```python
afinn_dict= aff.set_index('word')['value'].to_dict()
```

```python
type(afinn_dict)
```

```python
def sentiment_count(text:str):
  sent_count=0
  if text:
    sentence= nlp(text)
    for token in sentence:
      sent_count+= afinn_dict.get(token.lemma_,0)
    return sent_count
```
