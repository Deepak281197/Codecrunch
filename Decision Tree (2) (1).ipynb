# Company Data

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,classification_report
import seaborn as sns
```

```python
# Import Dataset
data=pd.read_csv('Company_Data.csv')
```

```python
data
```

```python
data
```

```python
# conver the data into numeric
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc=LabelEncoder()
```

```python
data['ShelveLoc']=lab_enc.fit_transform(data['ShelveLoc'])
```

```python
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
target=data['ShelveLoc']
```

```python
target
```

```python
data=data.drop(['ShelveLoc','Age','Education','Urban','US'],axis=1)
```

```python
features=data
```

```python
features
```

```python
# Visualization
sns.pairplot(data=data)
```

```python
# split the data using traintest_split
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.7,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec_tree=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=2)
```

```python
dec_tree.fit(x_train,y_train)
```

```python
y_pred=dec_tree.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn import tree
```

```python
tree.plot_tree(dec_tree);
```

# Fraud Check

```python
# Import dataset
df1=pd.read_csv('Fraud_check.csv')
```

```python
df1
```

```python
# Split the data into two types using lamda
df1['Taxable.Income'] = df1['Taxable.Income'].apply(lambda x: 'Risky' if x <= 30000 else 'Good')
```

```python
df1
```

```python
df1=pd.get_dummies(df1,columns=['Undergrad','Marital.Status','Urban'],drop_first=True)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df1
```

```python
# let's plot pair plot to visualise the attributes all at once
sns.pairplot(data=df1, hue = 'Taxable.Income')
```

```python
target=df1['Taxable.Income']
```

```python
target
```

```python
features=df1.drop('Taxable.Income',axis=1)
```

```python
features
```

```python
# Split the data into traintest_split
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.7,random_state=50)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Use model to predict
dec_tree=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=8)
```

```python
dec_tree.fit(x_train,y_train)
```

```python
y_pred=dec_tree.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Plot the decision tree
tree.plot_tree(dec_tree);
```
