```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
kdata=load_iris()
```

```python
kdata
```

```python
features=pd.DataFrame(kdata.data, columns=kdata.feature_names)
```

```python
features
```

```python
target=pd.DataFrame(kdata.target,columns=['Target'])
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.75,random_state=100)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
conffusion_mat=confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(conffusion_mat,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

```python
score=y_test.value_counts()
```

```python
score
```
