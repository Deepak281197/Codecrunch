# Aprioro & Association

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=pd.read_csv('Titanic.csv')
```

```python
data
```

```python
data.Gender.unique()
```

```python
data.Age.unique()
```

```python
data.Survived.unique()
```

```python
data.Class.unique()
```

```python
new_data=pd.get_dummies(data)
```

```python
new_data
```

```python
apriori=apriori(new_data,min_support=0.5,use_colnames=True)
```

```python
apriori
```

```python
data.shape
```

```python
data.Gender.value_counts()
```

```python
association_rules(apriori,metric='confidence',min_threshold=0.5,support_only=False)
```
