## Crosstab

```python
df1.columns
```

```python
pd.crosstab(index=df1['name_of_car'], columns=[df1['EngineSize'],df1['Weight']])
```

```python
pd.crosstab(index=[df1['name_of_car'],df1['Origin']], columns=[df1['EngineSize'],df1['Weight']],margins=True)
```

### Groupby

```python
df1['Origin'].value_counts()
```

```python
df1.groupby(['name_of_car']).get_group('Toyota')
```

```python
import numpy as np
df1.groupby([df1['name_of_car'],df1['Origin'],df1['Type']])['Weight','EngineSize'].agg([max,min,np.mean])
```

```python
df1[['Weight','EngineSize']].groupby([df1['name_of_car'],df1['Origin'],df1['Type']]).agg([max,min,np.mean])
```
